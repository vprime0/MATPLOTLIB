```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


gas=pd.read_csv("gas_prices.csv")
plt.figure(figsize=(9,5))
plt.plot(gas["Year"] ,gas["USA"],"b.-",label="USA")
plt.plot(gas["Year"],gas["Canada"],"r.-",label="CANADA")
plt.plot(gas["Year"],gas["South Korea"],label="South Korea",marker=".")
plt.xticks(gas["Year"][::3])
plt.legend()
```

```python
plt.title("GRAPH",fontdict={"fontweight":"bold","fontsize":20})
countries=["USA","Canada","Japan"]

for c in countries:
    plt.plot(gas["Year"],gas[countries],marker=".")
    
plt.xticks(gas["Year"][::3].tolist()+[2011]+[2015])
plt.savefig("gas_fig",dip=300)
```
